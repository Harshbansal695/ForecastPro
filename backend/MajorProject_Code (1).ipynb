```python
order_payments = data4.groupby('order_id')['payment_value'].sum().reset_index()
df = pd.merge(data6[['order_id', 'order_purchase_timestamp']], order_payments, on='order_id')
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df.set_index('order_purchase_timestamp', inplace=True)
```

```python
ts = df['payment_value'].resample('W').sum()
ts = ts['2017-01-01':'2018-08-15']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(ts, color='navy', lw=2)
plt.title('Weekly Sales Revenue (2017-2018)')
plt.xlabel('Date')
plt.ylabel('Revenue (BRL)')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import warnings
```

```python
result = adfuller(ts)
print(f'ADF Statistic: {result[0]:.4f}')
print(f'p-value: {result[1]:.4f} (Non-stationary if > 0.05)')
```

```python
train_size = len(ts) - 12
train, test = ts[0:train_size], ts[train_size:]
```

```python
model_arima = ARIMA(train, order=(1, 1, 1))
arima_results = model_arima.fit()
arima_forecast = arima_results.forecast(steps=12)
```

```python
scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(train.values.reshape(-1, 1))
```

```python
n_input = 4
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
model_lstm = Sequential([
    LSTM(64, activation='relu', input_shape=(n_input, 1), return_sequences=True),
    LSTM(32, activation='relu'),
    Dense(1)
])
model_lstm.compile(optimizer='adam', loss='mse')
model_lstm.fit(generator, epochs=50, verbose=0)
```

```python
lstm_preds = []
batch = scaled_train[-n_input:].reshape((1, n_input, 1))
for _ in range(len(test)):
    pred = model_lstm.predict(batch)[0]
    lstm_preds.append(pred)
    batch = np.append(batch[:, 1:, :], [[pred]], axis=1)

lstm_forecast = scaler.inverse_transform(lstm_preds).flatten()
```

```python
def evaluate(actual, forecast, name):
    rmse = np.sqrt(mean_squared_error(actual, forecast))
    mae = mean_absolute_error(actual, forecast)
    print(f"{name} Results -> RMSE: {rmse:.2f} | MAE: {mae:.2f}")
    return rmse
arima_rmse = evaluate(test, arima_forecast, "ARIMA")
lstm_rmse = evaluate(test, lstm_forecast, "LSTM")
```

```python
plt.figure(figsize=(14, 7))
plt.plot(train.index, train, label='Historical Data')
plt.plot(test.index, test, label='Actual Sales (Test)')
plt.plot(test.index, arima_forecast, label=f'ARIMA Forecast (RMSE: {arima_rmse:.0f})', linestyle='--')
plt.plot(test.index, lstm_forecast, label=f'LSTM Forecast (RMSE: {lstm_rmse:.0f})', linestyle=':')
plt.title('Sales Forecast Comparison: ARIMA vs LSTM')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts, model='additive', period=12)
fig = decomposition.plot()
fig.set_size_inches(14, 10)
plt.suptitle('Weekly Sales Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
arima_errors = test.values - arima_forecast
lstm_errors = test.values - lstm_forecast
sns.kdeplot(arima_errors, label='ARIMA Errors', fill=True, color="red")
sns.kdeplot(lstm_errors, label='LSTM Errors', fill=True, color="green")
plt.title('Error Distribution Comparison')
plt.xlabel('Residual Value')
plt.legend()
plt.show()
```
